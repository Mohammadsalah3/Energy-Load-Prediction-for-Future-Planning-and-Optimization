```python

from sklearn.impute import KNNImputer

# List of features to modify
features_to_modify = [
    'Lighting Consumption (kWh)', 'HVAC Consumption (kWh)',
    'Carbon Emission Rate (g CO2/kWh)', 'Building Age (years)',
    'Equipment Age (years)', 'Building Size (m²)',
    'Insulation Quality Score', 'Local Energy Production (kWh)',
    'Grid Stability Score', 'Solar Irradiance (W/m²)',
    'Smart Plug Usage (kWh)', 'Water Usage (liters)',
    'Energy Savings Target (%)', 'Room-Level Energy Consumption (kWh)',
    'Zonal Heating/Cooling Data (kWh)', 'IoT Sensor Count',
    'Thermal Comfort Index', 'Energy Savings Potential (%)'
]

# Step 1: Replace negative values in specified columns with NaN
data[features_to_modify] = data[features_to_modify].applymap(lambda x: np.nan if x < 0 else x)

# Step 2: Sort by "Building Size" and "Building Type" if "Building Type" is categorical and doesn't affect imputation
data = data.sort_values(by=[ "Building Type","Building Size (m²)"])

# Step 3: Initialize KNN Imputer with 5 neighbors
imputer = KNNImputer(n_neighbors=5)

# Step 4: Apply KNN Imputer only to the specified columns (features_to_modify)
# Extract only the columns to modify, apply KNN, and keep other values intact
data_to_impute = data[features_to_modify]
imputed_data = imputer.fit_transform(data_to_impute)

# Step 5: Convert imputed data back to a DataFrame
imputed_data_df = pd.DataFrame(imputed_data, columns=features_to_modify)

# Step 6: Only replace NaN values in the original DataFrame, leaving non-NaN values intact
for col in features_to_modify:
    data[col] = data[col].combine_first(imputed_data_df[col])

# Now, the `data` DataFrame has NaN values imputed, while original non-NaN values remain unchanged
```

```python
data = data.sort_index()
```

```python
# Create a copy of the data
data_copy = data.copy()

# Convert Timestamp to datetime
data_copy['Timestamp'] = pd.to_datetime(data_copy['Timestamp'])

# Add a Date column
data_copy['Date'] = data_copy['Timestamp'].dt.date

# List of relevant features
features_to_average = [
    'Energy Consumption (kWh)', 'Temperature (°C)', 'Humidity (%)', 
    'Lighting Consumption (kWh)', 'HVAC Consumption (kWh)', 
    'Energy Price ($/kWh)', 'Carbon Emission Rate (g CO2/kWh)', 
    'Power Factor', 'Power Outage Indicator', 'Indoor Temperature (°C)', 
    'Building Age (years)', 'Equipment Age (years)', 'Energy Efficiency Rating', 
    'Building Size (m²)', 'Insulation Quality Score', 
    'Historical Energy Consumption (kWh)', 'Maintenance Status', 
    'Demand Response Participation', 'Local Energy Production (kWh)', 
    'Grid Stability Score', 'Solar Irradiance (W/m²)', 
    'Smart Plug Usage (kWh)', 'Water Usage (liters)', 
    'Energy Savings Target (%)', 'Room-Level Energy Consumption (kWh)', 
    'Zonal Heating/Cooling Data (kWh)', 'Electric Vehicle Charging Status', 
    'IoT Sensor Count', 'Thermal Comfort Index', 'Energy Savings Potential (%)', 
    'Peak Demand Reduction Indicator'
]

# List of non-numeric features
non_numeric_features = ['Building Orientation', 'Occupancy Schedule']

# Filter numeric columns only
numeric_features = [col for col in features_to_average if pd.api.types.is_numeric_dtype(data_copy[col])]

# Process daily averages for each building type
daily_data_frames = []
building_types = ['Residential', 'Commercial', 'Industrial']

for b_type in building_types:
    building_data = data_copy[data_copy['Building Type'] == b_type]

    # Calculate mean for numeric features
    daily_avg = building_data.groupby('Date')[numeric_features].mean()
    
    # Calculate mode for non-numeric features
    daily_mode = building_data.groupby('Date')[non_numeric_features].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
    
    # Combine results
    daily_result = pd.concat([daily_avg, daily_mode], axis=1).reset_index()
    daily_result['Building Type'] = b_type
    daily_data_frames.append(daily_result)

# Combine the results
daily_data = pd.concat(daily_data_frames, ignore_index=True)

# Add the Date as a Timestamp column
daily_data['Timestamp'] = pd.to_datetime(daily_data['Date'])

# Drop the original Date column
daily_data = daily_data.drop(columns=['Date'])

# Reorder columns
columns_to_include = ['Timestamp', 'Building Type'] + numeric_features + non_numeric_features
daily_data = daily_data[columns_to_include]

# Save the result to CSV
output_path = r"C:\Users\mah19\OneDrive\سطح المكتب\project\electricity_consumption_optimization_dataset_updated.csv"
daily_data.to_csv(output_path, index=False)


```

```python
# Compute weights inversely proportional to errors
rf_weight = 1 / rf_mse
xgb_weight = 1 / xgb_mse
total_weight = rf_weight + xgb_weight

# Weighted average of predictions
hybrid_preds = (rf_weight * rf_preds + xgb_weight * xgb_preds) / total_weight

# Evaluate hybrid model
hybrid_mse = mean_squared_error(y_test, hybrid_preds)
print("Random Forest MSE:", rf_mse)
print("XGBoost MSE:", xgb_mse)
print("Hybrid Model MSE:", hybrid_mse)
```
